```python
import string
from collections import Counter
```

```python
#CLEAN TEXT FUNCTION
def clean_text(text):
    return text.lower().translate(str.maketrans('', '', string.punctuation))
#COUNT SPECIFIC WORD
def count_specific_word(text, target_word):
    words = clean_text(text).split()
    return words.count(target_word.lower())
#MOST COMMON WORD
def most_common_word(text):
    words = clean_text(text).split()
    word_counts = Counter(words)
    return word_counts.most_common(1)[0]
#AVERAGE WORD LENGTH
def average_word_length(text):
    words = clean_text(text).split()
    total_length = sum(len(word) for word in words)
    return total_length / len(words) if words else 0
#COUNT PARAGRAPHS
def count_paragraphs(text):
    paragraphs = text.strip().split("\n\n")
    return len([p for p in paragraphs if p.strip() != ""])
#COUNT SENTENCES
def count_sentences(text):
    sentences = text.replace('!', '.').replace('?', '.').split('.')
    return len([s for s in sentences if s.strip() != ""])
#MAIN PROGRAM
with open('/content/News Article for Python Assessment.txt', 'r', encoding='utf-8') as f:
    article = f.read()
    word_to_check = "the"
    print("Count of specific word:", count_specific_word(article, word_to_check))
    print("Most common word:", most_common_word(article))
    print("Average word length:", round(average_word_length(article), 2))
    print("Number of paragraphs:", count_paragraphs(article))
    print("Number of sentences:", count_sentences(article))
```
